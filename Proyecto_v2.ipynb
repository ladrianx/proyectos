Cambiamos los datos *chr* a *factor*

```python
%%R
chr_to_factor <- function(col) {
    if(is.character(col))
    { as.factor(col) }
    else { col }
}
df_ckd_convert <- lapply(df_ckd,chr_to_factor)
df_ckd <- data.frame(df_ckd_convert)

df_ckd2_convert <- lapply(df_ckd2,chr_to_factor)
df_ckd2 <- data.frame(df_ckd2_convert)
```

Se intenta respetar el tipo de variable que se describe en la base de datos, pero al cambiar las variables indicadas al tipo *factor* tienen una alta cardinalidad.

Sin embargo, los valores son tan poco frecuentes, como en el caso de *age* que al aplicar el método *RareLabelEncoder* todos los datos son agrupados en la clasificación *rare*

Por está razón, lo más conveniente parece considerar estas variables como numéricas continuas

Definamos las variables ordinales tipo *factor*

```python
%%R
df_ckd2$sg <- as.factor(df_ckd2$sg)
df_ckd2$al <- as.factor(df_ckd2$al)
df_ckd2$su <- as.factor(df_ckd2$su)
```
